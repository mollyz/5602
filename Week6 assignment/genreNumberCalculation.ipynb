```python
import csv  
import json  
```

```python
# Open the CSV  
f = open( './albumlist.csv', 'rU' ,encoding = "ISO-8859-1")  
# Change each fieldname to the appropriate field name. I know, so difficult.  
reader = csv.DictReader( f)  
# Parse the CSV into JSON  
print("JSON parsed!")


# find all genres
genres = []
for row in reader:
    cur_dict = dict(row)
    del cur_dict['Number']
    
    if cur_dict['Genre'] not in genres:
        genres.append(cur_dict['Genre'])
```

```python
genres
```

```python
# build dict
data = {}
for row in reader:
    cur_dict = dict(row)
    
    year = cur_dict['Year']
    if cur_dict['Year'] in data.keys():
        data[year][cur_dict['Genre']] += 1
    else:
        # . .....
        data[year] = {"year": year}
        for each in genres:
            data[year][each] = 0
        data[year][cur_dict['Genre']] = 1

# final output
output = []
for key in data.keys():
    output.append(data[key])
```
